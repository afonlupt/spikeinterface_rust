```python
import numpy as np
import spikeinterface.full as si
```

```python
def generate_data_num_channels(num_channels=128,seed=42):
    si.set_global_job_kwargs(n_jobs=0.5)
    static,sorting = si.generate_ground_truth_recording(durations=[50],num_channels=num_channels,sampling_frequency=25000.0,num_units= 100, seed=seed)
    print(sorting)
    return static, sorting

def save_data(recording,folder):
    recording.save_to_folder(folder = folder,overwrite=True)

def save_data_memory(recording):
    res = recording.save_to_memory()
    return res
```

```python
recording, sorting = generate_data_num_channels()
```

```python
recording.save_to_folder(folder = "../data/128_channels/recording",overwrite=True)
sorting.save_to_folder(folder = "../data/128_channels/sorting",overwrite=True)
np.save("../data/128_channels/templates.npy", recording.templates)
```
